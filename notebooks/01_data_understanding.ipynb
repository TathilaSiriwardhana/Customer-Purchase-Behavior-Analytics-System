## Dataset Overview

This dataset contains demographic information and purchase behavior of customers.
Each row represents a single customer.

### Columns:
- CustomerID: Unique identifier for each customer
- Gender: Gender of the customer
- Age: Age of the customer
- EstimatedSalary: Approximate annual salary
- Purchased: Indicates whether the customer made a purchase (1 = Yes, 0 = No)
